```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path =r"C:\Users\vaish\Downloads\archive (2)\Unemployment in India.csv"
df = pd.read_csv(r"C:\Users\vaish\Downloads\archive (2)\Unemployment in India.csv")
```

```python
print("Column Names:", df.columns)
```

```python
df.columns = df.columns.str.strip()
```

```python
print("\nFirst 5 rows of the dataset:")
print(df.head())
```

```python
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
df = df.dropna()
```

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
```

```python
print("\nBasic Statistics:")
print(df.describe())
```

```python
 plt.figure(figsize=(12, 6))
 sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=df)
 plt.title('Unemployment Rate Over Time')
 plt.xlabel('Date')
 plt.ylabel('Unemployment Rate (%)')
 plt.grid(True)
 plt.show()
```

```python
 if 'Region' in df.columns:
     plt.figure(figsize=(14, 7))
     sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', hue='Region',data=df)
     plt.title('Unemployment Rate by Region Over Time')
     plt.xlabel('Date')
     plt.ylabel('Unemployment Rate (%)')
     plt.legend(title='Region')
     plt.grid(True)
     plt.show()
```
